```python
import pandas as pd
import numpy as np
```

```python
songs_df = pd.read_csv('Data/data.csv')
songs_df.head()
```

```python
songs_df.columns
```

```python
#Rearranging the column orders
column_names = ['id','name', 'artists', 'release_date', 'year', 'explicit','popularity', 'duration_ms', 'acousticness', 'danceability', 
                'energy', 'instrumentalness', 'liveness', 'loudness',
               'mode', 'key', 'speechiness', 'tempo', 'valence']
songs_df = songs_df.reindex(columns=column_names)
songs_df.head()
```

```python
#Stripping characters from the artists column
songs_df['artists'] = songs_df.artists.str.strip("['']")
songs_df['artists'] = songs_df.artists.str.strip('""')
songs_df.head(20)
```

```python
songs_df.info()
```

```python
songs_df.shape
```

```python
#Used To find artists in the song dataframe
songs_df[songs_df.artists == 'Drake']
```

```python
#Number of unique artists in the Song Data Frame
len(songs_df.artists.unique())
```

```python
#Reading in the artist with genres csv
artist_genre = pd.read_csv('Data/data_w_genres.csv')
artist_genre
```

```python
#Merginf the songs and artists dataframes to get the genres to for songs
songs_w_genres_df = songs_df.merge(artist_genre, on='artists', how='left')
songs_w_genres_df.head()
```

```python
songs_w_genres_df.columns
```

```python
#Dropping the duplicated columns from the merge
songs_w_genres_df.drop(['release_date','acousticness_y',
       'danceability_y', 'duration_ms_y', 'energy_y', 'instrumentalness_y',
       'liveness_y', 'loudness_y', 'speechiness_y', 'tempo_y', 'valence_y',
       'popularity_y', 'key_y', 'mode_y', 'count'], axis=1, inplace=True)
songs_w_genres_df.head()
```

```python
#Renaming the Columns by stipping off _x
songs_w_genres_df.rename(columns=lambda x:x.strip('_x'), inplace=True)
songs_w_genres_df.head()
```

```python
#setting the index to be the id of the songs
songs_w_genres_df.set_index('id', inplace=True)
songs_w_genres_df.head()
```
